```python
#Silhouette Hierachical

from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

iris = datasets.load_iris()
X = iris.data
n_clusters = 5

ac = AgglomerativeClustering(n_clusters, linkage='ward')

cluster_labels = ac.fit_predict(X)

# Create a scatter plot
plt.figure(figsize=(8, 6))

# Plot data points with colors based on their cluster assignments
for i in range(n_clusters):
    plt.scatter(X[cluster_labels == i, 0], X[cluster_labels == i, 1], label=f'Cluster {i + 1}')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Hierarchical Clustering (Agglomerative Clustering)")
plt.legend()
plt.show()

score = silhouette_score(X, cluster_labels, metric='euclidean')

print('Silhouette Score: %.3f' % score)
```

```python
#Silhouette Spectral

from sklearn import datasets
from sklearn.cluster import SpectralClustering
from sklearn.metrics import silhouette_score

iris = datasets.load_iris()
X = iris.data
n_clusters = 5

spectral_clustering = SpectralClustering(n_clusters, eigen_solver='arpack', affinity='rbf', n_init=100)

cluster_labels = spectral_clustering.fit_predict(X)

# Create a scatter plot
plt.figure(figsize=(8, 6))

# Plot data points with colors based on their cluster assignments
for i in range(n_clusters):
    plt.scatter(X[cluster_labels == i, 0], X[cluster_labels == i, 1], label=f'Cluster {i + 1}')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Spectral Clustering")
plt.legend()
plt.show()

score = silhouette_score(X, cluster_labels, metric='euclidean')

print('Silhouette Score: %.3f' % score)
```

```python
#Calinski-Harabasz Index K-Means

from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score

iris = datasets.load_iris()
X = iris.data

# Specify the number of clusters you want
n_clusters = 5

kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)

cluster_labels = kmeans.fit_predict(X)

score = calinski_harabasz_score(X, cluster_labels)

print('Calinski-Harabasz Index: %.3f' % score)
```

```python
#Rand Index K-Means

from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

iris = datasets.load_iris()
X = iris.data
ground_truth_labels = iris.target  # Ground truth labels

km = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=0)
cluster_labels = km.fit_predict(X)

rand_index = adjusted_rand_score(ground_truth_labels, cluster_labels)

print('Rand Index: %.3f' % rand_index)
```

```python
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import normalized_mutual_info_score

iris = datasets.load_iris()
X = iris.data
true_labels = iris.target  # Ground truth labels

# Specify the number of clusters you want
n_clusters = 5

kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)

cluster_labels = kmeans.fit_predict(X)

nmi = normalized_mutual_info_score(true_labels, cluster_labels)

print('Normalized Mutual Information (NMI): %.3f' % nmi)
```
